```python
data=pd.read_csv("C:/Users/cuiyao/毕业设计SVM/bwz_snr.csv",index_col=0)
X=data.iloc[:,0:120076] #  分割样本特征
Y=data.iloc[:,12006:12007]  #  分割样本标签
y=np.ravel(Y) 
data
```

```python
result={"linear_acu_test":[],
            "poly_acu_test":[],
            "rbf_acu_test":[]}
for i in range(1,101,1):
    random_state = np.random.RandomState(i)
    x_train,x_test,y_train,y_test=train_test_split(X,y,test_size=0.3,random_state=i)  #  分割训练集和测试集，测试集占0.4
    model=svm.SVC(kernel="linear",probability=True,decision_function_shape="ovo")  #  定义模型，线性核函数，ovo分类方法
    acu_test=model.fit(x_train,y_train).score(x_test,y_test)
    result["linear_acu_test"].append(acu_test)
    model=svm.SVC(kernel="poly",probability=True,decision_function_shape="ovo")  #  定义模型，线性核函数，ovo分类方法
    acu_test=model.fit(x_train,y_train).score(x_test,y_test)
    result["poly_acu_test"].append(acu_test)
    model=svm.SVC(kernel="rbf",probability=True,decision_function_shape="ovo")  #  定义模型，线性核函数，ovo分类方法
    acu_test=model.fit(x_train,y_train).score(x_test,y_test)
    result["rbf_acu_test"].append(acu_test)
result=pd.DataFrame(result)
```
